```python
import pandas as pd

# Define the file paths
file_path1 = 'registry/distribution.txt'
file_path2 = 'registry/speciesprofile.txt'
file_path3 = 'registry/taxon.txt'

# Read the text files into DataFrames
df1 = pd.read_csv(file_path1, delimiter='\t')
df2 = pd.read_csv(file_path2, delimiter='\t')
df3 = pd.read_csv(file_path3, delimiter='\t')

# Merge the DataFrames on the 'id' column
merged_df = pd.merge(df1, df2, on='id', how='outer')
merged_df = pd.merge(merged_df, df3, on='id', how='outer')

# Display the merged DataFrame
print(merged_df)
```

```python
plantae_count = (merged_df['kingdom'] == 'Plantae').sum()
print(plantae_count)
```

```python
# Filter the DataFrame to only include rows where 'kingdom' is 'Plantae'
plantae_df = merged_df[merged_df['kingdom'] == 'Plantae']

# Display the filtered DataFrame
print(plantae_df)

# Save the merged DataFrame to a CSV file
plantae_df.to_csv('plantae.csv', index=False)
```

```python
for column in plantae_df.columns:
    print(f"Value counts for column '{column}':")
    print(plantae_df[column].value_counts())
    print("\n")
```

```python
columns_to_remove = ['taxonID', 'locationID', 'countryCode', 'occurrenceStatus', 'taxonRank']

# Create a new dataframe excluding the unwanted columns
new_plantae_df = plantae_df.drop(columns_to_remove, axis=1)

# You can now use the new_plantae_df which has the excluded columns removed.
print(new_plantae_df)
```
